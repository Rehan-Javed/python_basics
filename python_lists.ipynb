```python
# This is a list of strings where

letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
```

```python
letters[0], letters[4], letters[-1]
```

```python
for letter in letters:
        # Each of the strings is one character
        print(len(letter))

        # Each of the strings is a letter
        print(letter.isalpha())
```

### General slicing format is letters [ start : stop : step ]



```python
print("Select from the second element to the end")
print(letters[1:])
```

```python
print("Select from the start, stop before the last element")
print(letters[:-1])
```

```python
print("Start at index 1, stop 2 elements before the end")
print(letters[1:-2])
```

```python
print("From index 0 to before 9, take every 2nd element")
print(letters[0:9:2])

print("Take every 2nd element from start to end")
print(letters[::2])

print("Reverse the list and take every 2nd element")
print(letters[::-2])

print("Reverse the list completely")
print(letters[::-1])
```
