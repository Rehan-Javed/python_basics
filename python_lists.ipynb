What’s the difference between shallow and deep copy?

```python
# This works when you have lists of lists

# make a list  x
x = [[1,3,5,7,9], [2,4,6,8,10]]

# Make shallow copy of thex as y 
# The inner lists are shared
y = x.copy()         


y[0].append(11)


import copy
# Make deepcopy of the list
# The inner lists are only copied but they are not updated
z = copy.deepcopy(x) 

# Appending to y also appends to x
y[0].append(13)

# Appending to z doesnot append to x
z[0].append(15)
```

```python
print("x is ", x)
print("y is ", y)
print("z is ", z)
```

How do I safely delete while iterating?

```python
# Don’t remove from the same list while looping forward. Use a filtered rebuild:

x = [1, 2, 3, 4, 5, 6, 7, 8]

# keep evens -> [2,4]
x = [v for v in x if v % 2 == 0]  

x
```

How do I do a stable multi-key sort?

```python
# Sort by secondary key first, then primary key (Python sort is stable)

items = [("a", 2), ("b", 1), ("a", 1)]
items.sort(key=lambda t: t[1])  # secondary
print(items)
items.sort(key=lambda t: t[0])  # primary
print(items)

```

How do I remove duplicates but keep order?

```python
x = [3, 1, 3, 2, 1]

seen = set()
out = []
for v in x:
    print("now v is", v)
    if v not in seen:
        seen.add(v)
        out.append(v)
        print("seen :",seen)
        print("out :",out)
```

```python
x = [3, 1, 3, 2, 1]

seen = set()
out = []
for v in x:
    if v not in seen:
        seen.add(v)
        out.append(v)
print(out)
```

How do I use a list as a stack vs queue?

```python
# Stack (LIFO) (last in first out) is great:

stack = []
stack.append(1)
stack.append(2)
stack.append(3)
print(stack)
stack.pop()

stack
```

```python
# Queue (FIFO) with list is slow if you pop(0) a lot. Prefer collections.deque:

from collections import deque

q = deque([1,2,3])
q.append(4)
print(q)
q.popleft()
print(q)
```
